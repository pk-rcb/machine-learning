```python
# importing modules
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
#analysing the data and preprocessing
heartdata=pd.read_csv("/content/sample_data/heart_disease.csv");
```

```python
X = heartdata.drop(columns=['Heart Disease Status'])
y = heartdata['Heart Disease Status']
```

```python
categorical_cols = ['Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease', 'Diabetes',
                   'High Blood Pressure', 'Low HDL Cholesterol', 'High LDL Cholesterol', 'Alcohol Consumption',
                   'Stress Level', 'Sugar Consumption']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols),
        ('num', StandardScaler(), X.select_dtypes(include=['int64', 'float64']).columns)
    ])
```

```python
X_median = X.copy()
num_cols = X_median.select_dtypes(include=['int64', 'float64']).columns

# Impute numerical columns with median
median_imputer = SimpleImputer(strategy='median')
X_median[num_cols] = median_imputer.fit_transform(X_median[num_cols])

# Impute categorical columns with most frequent value
mode_imputer = SimpleImputer(strategy='most_frequent')
X_median[categorical_cols] = mode_imputer.fit_transform(X_median[categorical_cols])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_median, y, test_size=0.2, random_state=42)
```

```python
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_preprocessed, y_train)
```

```python
model = Pipeline(steps=[
    ('classifier', SVC(kernel='linear', random_state=42))
])
```

```python
model.fit(X_train_resampled, y_train_resampled)
```

```python
y_pred = model.predict(X_test_preprocessed)
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```
